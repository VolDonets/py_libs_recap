# Try the PyMuPDF lib

```python
import pymupdf
```

# Test simple text PDFs

```python
_uri = "../data/__test_pdfs/simple-text.pdf"
```

```python
doc = pymupdf.open(_uri) # open a document

for p_no, page in enumerate(doc): # iterate the document pages
    text = page.get_text() # get plain text (is in UTF-8)
    print(f"\n\nPage {p_no + 1} *** \n\n")
    print(text) # write text of page
```

# Test only one table PDFs

```python
_uri = "../data/__test_pdfs/table.pdf"
```

```python
import pymupdf
from pprint import pprint

doc = pymupdf.open(_uri) # open document
for p_no, page in enumerate(doc):
    tabs = page.find_tables() # locate and extract any tables on page
    print(f"{len(tabs.tables)} found on {p_no}") # display number of found tables

    if tabs.tables:  # at least one table found?
        pprint(tabs[0].extract())  # print content of first table
```
